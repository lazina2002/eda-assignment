```python
#11Identify and remove outliers in the km_driven column using the IQR method

Q1 = df["km_driven"].quantile(0.25)
Q3 = df["km_driven"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["km_driven"] < lower_bound) | (df["km_driven"] > upper_bound)]
print(f"Number of outliers detected: {outliers.shape[0]}")

df_cleaned = df[(df["km_driven"] >= lower_bound) & (df["km_driven"] <= upper_bound)]


print(f"New dataset shape after outlier removal: {df_cleaned.shape}")
```

```python
# 13What is the average depreciation in selling price based on the bike's age (current year - manufacturing
#year)D
current_year = 2023
df["bike_age"] = current_year - df["year"]

# Calculate average selling price for each bike age
avg_depreciation = df.groupby("bike_age")["selling_price"].mean().reset_index()

# Plot depreciation trend
plt.figure(figsize=(8, 5))
sns.lineplot(data=avg_depreciation, x="bike_age", y="selling_price", marker="o")
plt.title("Average Depreciation in Selling Price by Bike Age")
plt.xlabel("Bike Age (Years)")
plt.ylabel("Average Selling Price (â‚¹)")
plt.grid(True)
plt.show()

# Display the depreciation table
print(avg_depreciation)
```

```python
#insight>The average depreciation in selling price is calculated based on the bike's age (i.e., current year - manufacturing year).
#Newer bikes (5 years old) have an average selling price of â‚¹183,333, while older bikes (20+ years old) sell for much less.
#The depreciation trend follows a gradual decline, with some fluctuations due to high-value vintage bikes
```

```python
#15 Develop a correlation matrix for numeric columns and visualize it using a heatmap.
numeric_df = df.select_dtypes(include=['number'])

# Compute the correlation matrix
corr_matrix = numeric_df.corr()

# Set figure size
plt.figure(figsize=(10, 6))

# Create heatmap
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Title of the heatmap
plt.title("Correlation Matrix Heatmap", fontsize=14)

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Average Selling Price per Dealer
dealer_avg_price = df1.groupby("Dealer_Name")["Price ($)"].mean().reset_index()
plt.figure(figsize=(12, 5))
sns.barplot(x="Dealer_Name", y="Price ($)", data=dealer_avg_price)
plt.xticks(rotation=90)
plt.title("Average Selling Price per Dealer")
plt.show()
```

```python
# 2. Car Brand with Highest Price Variation
brand_price_variation = df1.groupby("Company")["Price ($)"].std().reset_index().sort_values(by="Price ($)", ascending=False)
print(brand_price_variation.head(1))
```

```python
# 3. Distribution of Car Prices for Each Transmission Type
plt.figure(figsize=(8, 5))
sns.boxplot(x="Transmission", y="Price ($)", data=df1)
plt.title("Distribution of Car Prices by Transmission Type")
plt.show()
```

```python
# 4. Distribution of Car Prices Across Regions
plt.figure(figsize=(12, 5))
sns.boxplot(x="Dealer_Region", y="Price ($)", data=df1)
plt.xticks(rotation=90)
plt.title("Car Price Distribution Across Regions")
plt.show()
```

```python
# 5. Distribution of Cars Based on Body Styles
body_style_counts = df1["Body Style"].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=body_style_counts.index, y=body_style_counts.values)
plt.xticks(rotation=90)
plt.title("Distribution of Cars Based on Body Styles")
plt.show()
```

```python
# 6. Average Selling Price by Customer Gender and Income
customer_price = df1.groupby(["Gender", "Annual Income"])['Price ($)'].mean().reset_index()
plt.figure(figsize=(10, 5))
# Reshape the DataFrame using pivot
customer_price_pivot = customer_price.pivot(index="Gender", columns="Annual Income", values="Price ($)")
sns.heatmap(customer_price_pivot, annot=True, cmap="coolwarm")
plt.title("Average Selling Price by Gender and Income")
plt.show()
```

```python
# 7. Car Prices and Number of Cars Sold by Region
region_price = df1.groupby("Dealer_Region")["Price ($)"].mean().reset_index()
region_sales = df1["Dealer_Region"].value_counts()
plt.figure(figsize=(12, 5))
sns.barplot(x=region_price["Dealer_Region"], y=region_price["Price ($)"])
plt.xticks(rotation=90)
plt.title("Average Car Price by Region")
plt.show()

plt.figure(figsize=(12, 5))
sns.barplot(x=region_sales.index, y=region_sales.values)
plt.xticks(rotation=90)
plt.title("Number of Cars Sold by Region")
plt.show()
```

```python
# 8. Car Price vs. Engine Size
plt.figure(figsize=(10, 5))
sns.scatterplot(x="Engine", y="Price ($)", data=df1)
plt.title("Car Price vs. Engine Size")
plt.show()
```

```python
# 9. Car Price Variation by Customer Income Bracket
plt.figure(figsize=(12, 5))
sns.boxplot(x="Annual Income", y="Price ($)", data=df1)
plt.xticks(rotation=90)
plt.title("Car Price Variation by Customer Income Bracket")
plt.show()
```

```python

# 10. Top 5 Car Models with Highest Sales
top_models = df1["Model"].value_counts().head(5).index
df1_top_models = df1[df1["Model"].isin(top_models)]
plt.figure(figsize=(12, 5))
sns.boxplot(x="Model", y="Price ($)", data=df1_top_models)
plt.xticks(rotation=90)
plt.title("Price Distribution of Top 5 Selling Car Models")
plt.show()
```

```python
# 11. Car Price vs. Engine Size Across Different Colors
plt.figure(figsize=(12, 5))
sns.boxplot(x="Color", y="Price ($)", data=df1)
plt.xticks(rotation=90)
plt.title("Car Price Variation Across Colors")
plt.show()
```

```python
# 12. Seasonal Trends in Car Sales
df1["Date"] = pd.to_datetime(df1["Date"])
df1["Month"] = df1["Date"].dt.month
plt.figure(figsize=(12, 5))
sns.lineplot(x="Month", y="Price ($)", data=df1)
plt.xticks(range(1, 13))
plt.title("Seasonal Trends in Car Sales")
plt.show()
```

```python
# importing all the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import re

# reading the data from the csv file
df = pd.read_csv('amazon.csv')

# Display basic information about the dataset
df.info()
df.head()

# Ensure that 'rating' and 'rating_count' are numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'].replace({'â‚¹': '', ',': ''}, regex=True), errors='coerce')

# Drop rows with NaN values in rating column
df = df.dropna(subset=['rating'])
```

```python
'''
Q1. What is the average rating for each product category?
'''
'''
Answer:-1
'''
average_rating_by_category = df.groupby('category')['rating'].mean()
print("Average Rating for Each Product Category:")
print(average_rating_by_category)
```

```python
'''
Q2. What are the top rating_count products by category?
'''
'''
Answer:-2
'''
top_rating_count_by_category = df.loc[df.groupby('category')['rating_count'].idxmax()][['category', 'product_name', 'rating_count']]
print("Top Rating Count Products by Category:")
print(top_rating_count_by_category)
```

```python
'''
Q3. What is the distribution of discounted prices vs. actual prices?
'''
'''
Answer:-3
'''
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='discounted_price', color='blue', label='Discounted Price', kde=True, bins=30)
sns.histplot(data=df, x='actual_price', color='orange', label='Actual Price', kde=True, bins=30)
plt.title('Distribution of Discounted Prices vs. Actual Prices')
plt.xlabel('Price (INR)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
'''
Q4. How does the average discount percentage vary across categories?
'''
'''
Answer:-4
'''

# Ensure that 'discount_percentage' is numeric
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

average_discount_percentage = df.groupby('category')['discount_percentage'].mean()
print("Average Discount Percentage Across Categories:")
print(average_discount_percentage)
```

```python
'''
Q5. What are the most popular product names?
'''
'''
Answer:-5
'''
most_popular_products = df['product_name'].value_counts().head(10)
print("Most Popular Product Names:")
print(most_popular_products)
```

```python
'''
Q6. What are the most popular product keywords?
'''
'''
Answer:-6
'''
# Extract keywords from product names
keywords = df['product_name'].str.cat(sep=' ').lower()
keywords_list = re.findall(r'\b\w+\b', keywords)
most_common_keywords = Counter(keywords_list).most_common(10)
print("Most Popular Product Keywords:")
print(most_common_keywords)
```

```python
'''
Q7. What are the most popular product reviews?
'''
'''
Answer:-7
'''
most_popular_reviews = df['review_title'].value_counts().head(10)
print("Most Popular Product Reviews:")
print(most_popular_reviews)
```

```python
'''
Q8. What is the correlation between discounted_price and rating?
'''
'''
Answer:-8
'''
df['discounted_price'] = df['discounted_price'].replace({'â‚¹': '', ',': ''}, regex=True).astype(float)
correlation = df[['discounted_price', 'rating']].corr().iloc[0, 1]
print(f"Correlation between Discounted Price and Rating: {correlation:.2f}")
```

```python
'''
Q9. What are the Top 5 categories based on the highest ratings?
'''
'''
Answer:-9
'''
top_categories_by_rating = df.groupby('category')['rating'].mean().nlargest(5)
print("Top 5 Categories Based on the Highest Ratings:")
print(top_categories_by_rating)
```

```python
'''
Q10. Identify any potential areas for improvement or optimization based on the data analysis.
'''
'''
Answer:-10
'''
# Example analysis: Identify categories with low ratings but high rating counts
low_rating_high_count = df[(df['rating'] < 3) & (df['rating_count'] > 100)]
print("Categories with Low Ratings but High Rating Counts:")
print(low_rating_high_count[['category', 'rating', 'rating_count']])
```

```python
# importing all the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# reading the data from the csv file
df = pd.read_csv('spotify.csv')

# Display basic information about the dataset
df.info()
df.head()
```

```python
'''
Q1. Read the dataframe, check null value if present then do the needful, check duplicate row, if present then do the needful.
'''
'''
Answer:-1
'''
# Read the dataframe
df = pd.read_csv('spotify.csv')

# Check for null values
print("Null values in each column:")
print(df.isnull().sum())

# Handle null values if present (e.g., drop or fill)
df.dropna(inplace=True)  # Dropping rows with null values

# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# Remove duplicate rows if present
if duplicates > 0:
    df.drop_duplicates(inplace=True)
```

```python
'''
Q2. What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram.
'''
'''
Answer:-2
'''
plt.figure(figsize=(10, 6))
sns.histplot(df['Popularity'], bins=30, kde=True)
plt.title('Distribution of Popularity Among Tracks')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
'''
Q3. Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot.
'''
'''
Answer:-3
'''
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Duration (ms)', y='Popularity')
plt.title('Relationship Between Popularity and Duration of Tracks')
plt.xlabel('Duration (ms)')
plt.ylabel('Popularity')
plt.show()
```

```python
'''
Q4. Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot.
'''
'''
Answer:-4
'''
artist_counts = df['Artist'].value_counts()
top_artist = artist_counts.idxmax()
top_artist_count = artist_counts.max()
print(f"Artist with the highest number of tracks: {top_artist} ({top_artist_count} tracks)")

# Display count of tracks for each artist using a countplot
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Artist', order=artist_counts.index)
plt.title('Count of Tracks for Each Artist')
plt.xlabel('Number of Tracks')
plt.ylabel('Artist')
plt.show()
```

```python
'''
Q5. What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each.
'''
'''
Answer:-5
'''
least_popular_tracks = df.nsmallest(5, 'Popularity')[['Artist', 'Track Name', 'Popularity']]
print("Top 5 Least Popular Tracks:")
print(least_popular_tracks)
```

```python
'''
Q6. Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist.
'''
'''
Answer:-6
'''
top_5_artists = artist_counts.nlargest(5).index
average_popularity = df[df['Artist'].isin(top_5_artists)].groupby('Artist')['Popularity'].mean()
print("Average Popularity for Each of the Top 5 Most Popular Artists:")
print(average_popularity)
```

```python
'''
Q7. For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.
'''
'''
Answer:-7
'''
for artist in top_5_artists:
    most_popular_track = df[df['Artist'] == artist].nlargest(1, 'Popularity')[['Track Name', 'Popularity']]
    print(f"Most Popular Track for {artist}:")
    print(most_popular_track)
```

```python
'''
Q8. Visualize relationships between multiple numerical variables simultaneously using a pair plot.
'''
'''
Answer:-8
'''
sns.pairplot(df[['Popularity', 'Duration (ms)']])
plt.title('Pair Plot of Popularity and Duration')
plt.show()
```

```python
'''
Q9. Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot.
'''
'''
Answer:-9
'''
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Artist', y='Duration (ms)')
plt.title('Duration of Tracks Across Different Artists')
plt.xticks(rotation=90)
plt.xlabel('Artist')
plt.ylabel('Duration (ms)')
plt.show()
```

```python
'''
Q10. How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot.
'''
'''
Answer:-10
'''
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Artist', y='Popularity')
plt.title('Distribution of Track Popularity for Different Artists')
plt.xticks(rotation=90)
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.show()
```
